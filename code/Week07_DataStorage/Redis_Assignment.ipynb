## How many users are there in the database?

```python
cursor = 0
user_count = 0

while True:
    cursor, keys = rd.scan(cursor=cursor, match='user:*:name', count=100)
    print(cursor, keys)
    user_count += len(keys)

    if cursor == 0:
        break

print(f"Number of users: {user_count}")
```

## What is the average number of follows per user?

```python
cursor = 0
follow_count = 0

while True:
    cursor, keys = rd.scan(cursor=cursor, match='user:*:follows', count=100)
    print(cursor, keys)
    for key in keys:
        follow_count += rd.scard(key)

    if cursor == 0:
        break

print(f"Average number of follows per user: {follow_count/user_count}")
```

## How many users follows between 5-10 users?

```python
cursor = 0
user_5_10 = 0

while True:
    cursor, keys = rd.scan(cursor=cursor, match='user:*:follows', count=100)
    print(cursor, keys)
    for key in keys:
        follows = rd.scard(key)
        if follows >= 5 and follows <= 10:
            user_5_10 += 1

    if cursor == 0:
        break

print(f"No. of users that follow 5-10 users: {user_5_10}")
```

## Which account has the most followers?

```python
cursor = 0
max_followers = 0
user_with_most_followers = None

while True:
    cursor, keys = rd.scan(cursor=cursor, match='user:*:followed_by', count=100)
    print(cursor, keys)
    for key in keys:
        follows = rd.scard(key)

        if follows > max_followers:
            max_followers = follows
            user_with_most_followers = key

    if cursor == 0:
        break

print(f"User with the most followers: {user_with_most_followers} with {max_followers} followers")
```
