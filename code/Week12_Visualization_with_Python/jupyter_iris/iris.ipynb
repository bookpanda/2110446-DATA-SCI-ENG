##  Pair Plot with Matplotlib

```python
import matplotlib.patches as mpatches

n = len(iris_features)
figure, axes = plt.subplots(n, n, figsize=(15, 15))

for i in range(n):
    for j in range(n):
        if i == j:
            axes[i, j].hist(iris[iris_features[i]], bins=10, edgecolor='black')
        else:
            for species, group in iris.groupby('Species'):
                axes[i, j].scatter(group[iris_features[j]], group[iris_features[i]],
                                color=color_map[species],
                                alpha=0.3, edgecolor=None,
                                label=species)
            axes[i, j].set_xlim([features_min[iris_features[j]]-0.5, features_max[iris_features[j]]+0.5])
            axes[i, j].set_ylim([features_min[iris_features[i]]-0.5, features_max[iris_features[i]]+0.5])
        axes[i, j].get_xaxis().set_visible(False)
        axes[i, j].get_yaxis().set_visible(False)
        if (i == n-1):
            axes[i, j].get_xaxis().set_visible(True)
            axes[i, j].set_xlabel(iris_features[j])
        if (j == 0):
            axes[i, j].get_yaxis().set_visible(True)
            axes[i, j].set_ylabel(iris_features[i])

rect = []
for key, value in color_map.items():
    rect.append(mpatches.Rectangle((0, 0), 1, 1, fc=value))
    
figure.legend(rect, iris_species)
figure.suptitle("Pair Plot of the Iris Dataset")

plt.show()
```

## KMeans

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
cluster = kmeans.fit_predict(iris.loc[:, 'SepalLengthCm':'PetalWidthCm'])
```

```python
cluster 
```

```python
iris['cluster'] = cluster
```

```python
print(kmeans.cluster_centers_)
```

```python
color_map = ['blue', 'green', 'orange', 'red']

for cluster, group in iris.groupby('cluster'):
    plt.scatter(group['PetalLengthCm'], group['SepalLengthCm'],
                color=color_map[cluster],
                alpha=0.3, edgecolor=None,
                label=cluster)
    plt.scatter(kmeans.cluster_centers_[cluster, 2], kmeans.cluster_centers_[cluster, 0], color=color_map[3])
plt.legend(frameon=True, title='cluster')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Petal Length vs. Sepal Length')
plt.show()
```

```python
import matplotlib.patches as mpatches

n = len(iris_features)
plt.figure(figsize=(15, 15))
color_map = ['blue', 'green', 'orange', 'red']

for i in range(n):
    for j in range(n):
        plt.subplot(n, n, i * n + j + 1)
        
        if i == j:
            plt.hist(iris[iris_features[i]], bins=10, edgecolor='black')
        else:
            for cluster, group in iris.groupby('cluster'):
                plt.scatter(group[iris_features[j]], group[iris_features[i]],
                            color=color_map[cluster],
                            alpha=0.3, edgecolor=None,
                            label=f"Cluster {cluster}" if i == 0 and j == 1 else "")
                # Assuming kmeans.cluster_centers_ is available and correct
                plt.scatter(kmeans.cluster_centers_[cluster, j], kmeans.cluster_centers_[cluster, i],
                            color=color_map[3], label='Centroids' if i == 0 and j == 1 else "")
            plt.xlim([features_min[iris_features[j]]-0.5, features_max[iris_features[j]]+0.5])
            plt.ylim([features_min[iris_features[i]]-0.5, features_max[iris_features[i]]+0.5])
        
        if i != n-1:
            plt.xticks([])
        if j != 0:
            plt.yticks([])
        if i == n-1:
            plt.xlabel(iris_features[j])
        if j == 0:
            plt.ylabel(iris_features[i])

patches = [mpatches.Patch(color=color_map[i], label=f'Cluster {i}') for i in range(len(color_map) - 1)]
patches.append(mpatches.Patch(color=color_map[-1], label='Centroids'))
plt.figlegend(handles=patches, loc='upper right', title='Legend')

plt.suptitle("Pair Plot of KMeans Clustering of Iris Dataset")

plt.show()

```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np

```

```python
# Define color map
color_map = ['blue', 'green', 'orange', 'red']

# Create subplots
n = len(iris_features)
fig = make_subplots(rows=n, cols=n, subplot_titles=[f'{feature}' for feature in iris_features])

for i, feature_i in enumerate(iris_features):
    for j, feature_j in enumerate(iris_features):
        if i == j:  # Diagonal - Histograms
            fig.add_trace(go.Histogram(x=iris[feature_i], marker_color='blue'), row=i+1, col=j+1)
        else:  # Scatter plots for each pair of features
            for cluster in iris['cluster'].unique():
                cluster_df = iris[iris['cluster'] == cluster]
                fig.add_trace(go.Scatter(x=cluster_df[feature_j], y=cluster_df[feature_i],
                                         mode='markers', marker_color=color_map[cluster], name=f'Cluster {cluster}',
                                         marker=dict(size=5, opacity=0.3)), row=i+1, col=j+1)
            # Add centroids
            fig.add_trace(go.Scatter(x=[kmeans.cluster_centers_[cluster][j]], y=[kmeans.cluster_centers_[cluster][i]],
                                     mode='markers', marker_symbol='x', marker_color=color_map[3], marker_size=10,
                                     showlegend=(i == 1 and j == 0), name='Centroids'), row=i+1, col=j+1)

        # Update axes visibility and labels
        if i != n-1: fig.update_xaxes(showticklabels=False, row=i+1, col=j+1)
        else: fig.update_xaxes(title_text=feature_j, row=i+1, col=j+1)
        
        if j != 0: fig.update_yaxes(showticklabels=False, row=i+1, col=j+1)
        else: fig.update_yaxes(title_text=feature_i, row=i+1, col=j+1)

# Update layout and show figure
fig.update_layout(height=1200, width=1200, title_text="Pair Plot of KMeans Clustering of Iris Dataset", showlegend=True)
fig.show()
```
